## Intro
Reasons for it are numerous and well described elsewhere, however practical tutorials for a complete Devbox-in-the-cloud are lacking.

Two main reasons are:
* avoid maintaining a piece of metal in your home
* compute per dollar in the cloud grows exponentially and outcompetes most personal GPU use-cases

Here are our requirements:

* competetive pricing
* easy jupyterlab login using our custom domain and Google for auth
* open project folder via VSCode for better IDE experience
* auto-shutdown, to save that :money:
* full access to `root` terminal, so we can setup all kinds of goodies, such as:
    * tensorboard
    * mlflow
